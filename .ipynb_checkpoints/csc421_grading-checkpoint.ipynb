```python
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook, output_file
from bokeh.models import SingleIntervalTicker, LinearAxis
```

```python
grades = pd.read_csv('data/csc421_fall_2021_midterm_grades.csv')
print(grades)
```

```python
output_notebook()
hist, edges = np.histogram(grades['grade'], density=False, bins=40)
p = figure(title='CSC421 midterm  - ' + str(np.mean(grades['grade'])), tools='', 
           background_fill_color="#fafafa", x_axis_type=None)
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
           fill_color="navy", line_color="white", alpha=0.5)


ticker = SingleIntervalTicker(interval=1, num_minor_ticks=2)
xaxis = LinearAxis(ticker=ticker)
p.add_layout(xaxis, 'below')
show(p)

output_file("csc421_midterm.html")
```

```python
hist, edges = np.histogram(grades['grade'] * 5, density=False, bins=20)
p = figure(title='CSC421 midterm - ' + str(np.mean(grades['grade'] * 5)), tools='', 
           background_fill_color="#fafafa",x_axis_type=None)
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
           fill_color="navy", line_color="white", alpha=0.5)

ticker = SingleIntervalTicker(interval=5, num_minor_ticks=1)
xaxis = LinearAxis(ticker=ticker)
p.add_layout(xaxis, 'below')

show(p)
```
